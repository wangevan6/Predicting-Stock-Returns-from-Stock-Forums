```python
# Define the Extraction Function
def get_company_name(text):
    client = OpenAI()
    response = client.chat.completions.create(
      model="gpt-4-turbo",
      response_format={ "type": "json_object" },
      messages=[
        {"role": "system", "content": "You are a helpful assistant designed to output JSON."},
        {"role": "user", "content": "Identify the main company's ticker mentioned in this text:\"" + text + "\""}
      ]
    )
    print(response.choices[0].message.content)
    # Parse the JSON-like string to a dictionary
    parsed_response = json.loads(response.choices[0].message.content)
    try:
        # First, try to extract using the key 'company_ticker'
        ticker_symbol = parsed_response["company_ticker"]
    except KeyError:
        # If the key 'company_ticker' doesn't exist, then try the key 'ticker'
        ticker_symbol = parsed_response.get("ticker", "Ticker not found")
    return ticker_symbol 
```

```python
def get_post_posn(text):
    client = OpenAI()
    
    response_inner = client.chat.completions.create(
      model="gpt-3.5-turbo",
      response_format={ "type": "json_object" },
      messages=[
        {"role": "system", "content": "You are a helpful assistant designed to output JSON."},
        {"role": "user", "content": "Based on the analysis of the provided post, return the stock position in a JSON format with 'Position' as the key and 'Long' or 'Short' as the value. With limited reasoning or additional text. \"" + text + "\""}
      ]
    )
    print(response_inner.choices[0].message.content)
    try:
        position = json.loads(response_inner.choices[0].message.content)["Position"]
        if not isinstance(position, str):
            return "Error: Position must be a single string value"
    except Exception as e:
        # Handle any other unexpected errors
        position = "Error: An unexpected issue occurred"
    print()
    return position
```
